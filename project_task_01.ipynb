## 4. Define the aggregation fucntion

```python
def agg_all_lanes(rows):
    # detector time interval
    dt = 30 # seconds

    # Extract volume, speed, and occupancy values for all lanes
    volumes = rows[['vol_1', 'vol_2', 'vol_3', 'vol_4', 'vol_5']]
    speeds = rows[['speed_1', 'speed_2', 'speed_3', 'speed_4', 'speed_5']]
    occupancies = rows[['occ_1', 'occ_2', 'occ_3', 'occ_4', 'occ_5']]

    # Sum the volume of all lanes, excluding missing values
    volume = volumes.sum(skipna=True)

    # Calculate the volume-weighted average speed, excluding missing values
    if volume == 0:
        speed = 0
    else:
        speed = pd.Series(speeds.values * volumes.values).sum(skipna=True) / volume

    # Option 1: Sum occupancy of all lanes, excluding missing values
    # occupancy = occupancies.sum(skipna=True)
    # Option 2: Average occupancy of all lanes, excluding missing values
    occupancy = occupancies.mean(skipna=True)


    return pd.Series([volume, speed, occupancy], index=['volume', 'speed', 'occupancy'])

# Test the function with selected station_ID (402005 or 405395 or 406760).
where = (us101sb_2019_10_19['station_ID'] == 405395) | (us101sb_2019_10_19['station_ID'] == 402005) | (us101sb_2019_10_19['station_ID'] == 406760)
test_df = us101sb_2019_10_19[where]
test_df[['volume', 'speed', 'occupancy']] = test_df.apply(agg_all_lanes, axis=1)
test_df
```

```python
def agg_by_minutes(df):

    # Convert 'date' column to datetime format
    df['date'] = pd.to_datetime(df['date'])

    # Select the columns: 'date', 'station_ID', 'volume', 'speed', 'occupancy'
    df = df[['date', 'station_ID', 'volume', 'speed', 'occupancy']]

    # Group the data by one minute intervals
    df = df.groupby(['station_ID', df['date'].dt.floor('T')]).agg({
        # Sum the volume 
        'volume': 'sum',
        # Calculate the volume-weighted average speed, excluding missing values 
        'speed': lambda s: np.average(s, weights=df.loc[s.index, 'volume']) if df.loc[s.index, 'volume'].sum() != 0 else 0,
        # Calculate the mean occupancy
        'occupancy': 'mean'
    })
    return df

# Test the function with selected station_ID
test_by_minute_df = agg_by_minutes(test_df)
test_by_minute_df.reset_index(inplace=True)
test_by_minute_df.sample(10)
```
